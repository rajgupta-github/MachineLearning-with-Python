## UNIVARIATE PLOTTING

### 1. Create a bar chart using "Type 1" and its frequency.

```python
pokemon.info()
```

```python
pokemon['Type 1'].value_counts()
```

```python
import pandas as pd
def create():
    pokemon['Type 1'].value_counts().plot.bar(figsize=(20,7), colormap='Dark2', fontsize=13, yticks=np.arange(0, 150, 10))
    plt.xlabel('Type 1', fontsize=20)
    plt.ylabel('Number of Items', fontsize=20)
    plt.title('Bar Chart showing the Number of Items in each Type 1 value', fontsize=20)
    return None
create()
```

### 2. Create a line chart using "HP" and its frequency.

```python
pokemon['HP'].value_counts()
```

```python
import pandas as pd
def create():
    pokemon['HP'].value_counts().sort_index(ascending=True).plot(kind='line', figsize=(20,7), color='red', 
                                                                 fontsize=15, linestyle='-')

    plt.xlabel('HP', fontsize=20)
    plt.ylabel('Count', fontsize=20)
    plt.title('Line Chart showing the Count for each HP value', fontsize=20)
    return None
create()
```

### 3. Create an area chart using "Total" and its frequency.

```python
import pandas as pd
def create():
    pokemon['Total'].value_counts().sort_index().plot(kind='area', figsize=(15,9), fontsize=20, color='purple')
    plt.xlabel('TOTAL', fontsize=20)
    plt.ylabel('Count', fontsize=20)
    plt.title('AREA Chart showing the Count for each TOTAL value', fontsize=20)
    return None
create()    
```

### 7. Create a stacked chart on "Legendary" and "Generation" columns, for stacking on "Attack" and "Defense" values.

```python
pokemon.head(1)
```

```python
import pandas as pd
def create():
    pokemon.groupby(['Legendary', 'Generation']).mean()[['Attack', 'Defense']].plot(kind='bar', figsize=(15, 7), fontsize=13, stacked=True)
    plt.xlabel('Legendary, Generation', fontsize=15)
    plt.ylabel('Mean', fontsize=15)
    plt.title('Stacked Bar Chart showing the Legendary and Generation for each Attack and Defense', fontsize=15)
    plt.legend(labels=('Attack', 'Defense'), fontsize=13)
    return None
create()
```

### 8. Create a bivariate plot line on "HP", "Attack", "Defense", "Sp.Atk", "Sp.Def", "Speed", grouped by "Generation".

```python
import pandas as pd
def create():
    pokemon_stats_by_generation = pokemon.groupby('Generation').mean()[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
    pokemon_stats_by_generation.plot.line(fontsize=10)
    return None
create()
```

### 9. Create a pairplot dropping "Name" and using "Legendary" as hue and implement a palette of your choice(size=2).
