## 1. Batch Gradient Descent (BGD/GD)

 - In **Batch Gradient Descent optimazation**, **parameters** are **updated** after **calculating the cost function** on the **entire dataset**. That means, to take **one step** in the gradient descent optimization **towards minimum**, the **entire dataset** is taken into account to **calculate the cost function**.
<br>
<br>
 - To take **another step towards the minimum, gradient descent optimization algorithm** should go through the **entire dataset again** to calculate the cost function.
<br>
<br>
 - This approach works **fine** for **small datasets.**
<br>
<br>
 - But, what **if** our **dataset** is **very large?**

<a id=section101></a>

### 1.1 Disadvantages of Batch Gradient Descent

 - Say you want to build a **ML model on India Census data**. This dataset will have **133 92 00 000 records**. Thats 133.92 crores records.
<br>
<br>
 - To move one step closer to the minimum of cost function, Batch Gradient Descent optimization algorithm calculates cost function on the entire dataset.
<br>
<br>
 - This characteristic of Batch Gradient Descent makes it **computationaly very expensive** which leads to consuption of **more time to converge to minimum.**
<br>
<br>
 - What if the optimization algorithm doesn't have to go through all data to take a step towards minimum of cost function.

<a id=section2></a>

## 2. Stochastic Gradient Descent (SGD)

 - In this variant of Gradient Descent, The optimization algorithm looks at **one record(random) at a time** to take **a step** in the gradinet descent.
<br>
<br>
 - So the algorithm has **already(right from the first instance itself) started** it's journey towards minimum of cost function **unlike Batch Gradient Descent** where it has to go through **entire dataset** to take **one single step.**
<br>
<br>

### 2.1 Advantages of Stochastic Gradient Descent
 - 1.Takes less time to converge to the minimum of COST FUNCITON

 - 2 . **Stochastic Gradient Descent doesn't** suffer from the problem of **getting trapped in local minima.**
 <br>
 <br>

<a id=section201></a>

### 2.2 Disadvantages of Stochastic Gradient Descent
 - **SGD** often **converges much faster** compared to **GD** **but** the **error function** is **not** as **well minimized** **as in the case of GD**. Often **in most cases**, **the close approximation** that **you get in SGD** for the **parameter** values **are enough** because **they reach the optimal values and keep oscillating there.**

## 3. Mini Batch Gradient Descent

 - For Batch Gradient Descent and Stochastic Gradient Descent, the entire data shoulb be loaded into RAM. If we come across a scenario where the size of the data is so big that the entire data can't be loaded on to RAM, then we can use Mini-Batch Gradient Descent optimization to train ML model.

### The Process:
- The training data is divided into batches
<img src="https://raw.githubusercontent.com/insaid2018/optimization-files/master/7.png" width="500", height="400">
   - First batch is taken.
   - Weights are initialized with random values.
   - Cost Function is calculated considering 1st batch of the data.
   - Weights are updated.
   - Then second batch of data is taken.
   - Cost Function is calculated considering 2nd batch of data with updated weights.
   - Weights are updated again.
   - Then third batch of data is taken.
   - The process continous for specified no.of epochs(no.of times the model sees the entire dataset).
    
**Batch size can affect the time taken to converge or even final accuracy(or Cost Function).
Right Batch size can be arrived using GridSearchCV.**

## 4. max_iter (Epochs)

 - **Number of** times the **optimization algorithm sees** the **entire dataset**.

**For example:**
 - If **max_iter = 100** in **Batch Gradient Descent Optimization**, the optimization algorithm **sees the entire dataset 100 times** and after each epoch, **one step** in gradient descent is taken towards the minimum of the cost function. In the end, **100 steps are taken in total.**
<br>
<br>
 - If **max_iter = 5 in Stochastic Gradient Descent Optimization**, the optimization algorithm **sees the entire dataset 5 times** but **a step is taken** for **every single record**. Let us say we have **a dataset of 1000 records**. So, in the end, **a total of 5*1000 = 5000 steps** are taken towards the minimum of the cost function.

Usually, **max_iter ranges from 1 to 10** for **Stochastic Gradient Descent.**

Function to remove one of the co-related columns

```python
def remove_collinear_features(x, threshold):
    '''
    Objective:
        Remove collinear features in a dataframe with a correlation coefficient
        greater than the threshold. Removing collinear features can help a model
        to generalize and improves the interpretability of the model.
        
    Inputs: 
        threshold: any features with correlations greater than this value are removed
    
    Output: 
        dataframe that contains only the non-highly-collinear features
    '''
    
    # Dont want to remove correlations between Energy Star Score
    y = x['Loan Status']
    x = x.drop('Loan Status',axis = 1)
    
    # Calculate the correlation matrix
    corr_matrix = x.corr()
    iters = range(len(corr_matrix.columns) - 1)
    drop_cols = []

    # Iterate through the correlation matrix and compare correlations
    for i in iters:
        for j in range(i):
            item = corr_matrix.iloc[j:(j+1), (i+1):(i+2)]
            col = item.columns
            row = item.index
            val = abs(item.values)
            
            # If correlation exceeds the threshold
            if val >= threshold or val <= -threshold:
                # Print the correlated features and the correlation value
                # print(col.values[0], "|", row.values[0], "|", round(val[0][0], 2))
                drop_cols.append(col.values[0])

    # Drop one of each pair of correlated columns
    drops = set(drop_cols)
    x = x.drop(drops,axis = 1)
    
    # Add the score back in to the data
    x['Loan Status'] = y
               
    return x
```

#### 10.2 Evaluating performance of SGDClassifier for no.of epochs ranging from 1 to 20

```python
%%time
test_accuracy_list_epochs = []
train_accuracy_list_epochs = []

for  i in range(1,21):
    
    sgd = SGDClassifier(max_iter=i,loss = 'log')
    sgd_model = sgd.fit(X_train,y_train)
    
    y_train_preds = sgd_model.predict(X_train)
    y_test_preds = sgd_model.predict(X_test)
    
    y_train_accuracy = accuracy_score(y_train,y_train_preds)
    y_test_accuracy = accuracy_score(y_test,y_test_preds)
    
    train_accuracy_list_epochs.append(y_train_accuracy)
    test_accuracy_list_epochs.append(y_test_accuracy)
```

#### Plotting Accuracy of the model against no.of epochs 

```python
plt.plot(range(1,21),train_accuracy_list_epochs,'-',label = 'train')
plt.plot(range(1,21),test_accuracy_list_epochs,'-',label = 'test')
plt.xticks(range(1,21))
plt.xlabel('No.of Epochs')
plt.ylabel('Accuracy')
plt.grid()
plt.legend()
```

We can infer from the graph that after 3 epochs there is no much increase in accuracy. So we can set max_iter = 3. More epochs means more computations.

## 11. Mini Batch Gradient Descent

In the **preceding example**, we used the **fit method**, which requires that you **preload all the training data into memory**. You can **train** the model **in successive steps** by using the **partial_fit method** instead, which **runs** a **single iteration** on the **provided data**, __then keeps it in memory and adjusts it when receiving new data.__

```python
%%time
sgd_mini_batch_model = SGDClassifier(loss = 'log')
improvements = list()
for z in range(100):
    sgd_mini_batch_model.partial_fit(X_train, y_train,classes = [1,0])
    improvements.append(accuracy_score(y_test, sgd_mini_batch_model.predict(X_test)))
```

Having kept track of the algorithm’s partial improvements during __100 iterations__ over the __same data__, you can produce a graph and understand how the improvements work as shown in the following code. __It’s important to note that you should use different data at each step which is not shown above. There is an example of implementing SGDClassifier on mini-batches of data at end of this notebook.__

```python
plt.plot(range(1,101),improvements[:100],'o--')
plt.xticks(range(1,100),rotation='vertical')
plt.xlabel('Iterations')
plt.ylabel('Test Accuracy')
plt.grid()
```

It is a good practice to shuffle the data after each epoch.

```python
%%time
import numpy as np
import pickle

#Dumping train data as binary data using pickle

with open('X_train.pickle','wb') as file:
    pickle.dump(X_train_clipped,file)

#Dumping test data as binary data using pickle

with open('y_train.pickle','wb') as file:
    pickle.dump(y_train_clipped,file)



sgd_mini_batch_model_1 = SGDClassifier(loss='log') 

#Dumping SGDClassifier as binary object using pickle

with open('sgd_partial_fit.pickle','wb') as file:
    pickle.dump(sgd_mini_batch_model_1,file)

n_iter = 20
improvements_1 = list()


for n in range(n_iter):
    
    #Reading train data (binary data) into X_train_loaded as numpy array
    
    with open('X_train.pickle','rb') as file:
        X_train_loaded = pickle.load(file)
    
    #Reading test data (binary data) into y_train_loaded as numpy array
    
    with open('y_train.pickle','rb') as file:
        y_train_loaded = pickle.load(file)
       
    #Reading SGDClassifier binary object into sgd_mini_batch_model_loaded
    
    with open('sgd_partial_fit.pickle','rb') as file:
        sgd_mini_batch_model_loaded = pickle.load(file)
    
    np.random.seed(n)
    np.random.shuffle(X_train_loaded)
    np.random.seed(n)
    np.random.shuffle(y_train_loaded)
    j= 0
    for i in range(0,798):
        sgd_mini_batch_model_loaded.partial_fit(X_train_clipped[j:j+100,],y_train_clipped[j:j+100],classes = [1,0])
        j = j+100
    improvements_1.append(accuracy_score(y_test, sgd_mini_batch_model_loaded.predict(X_test)))
    
    #Dumping SGDClassifier as binary object using pickle. This object is partially trained. The pickle file can 
    # again be loaded and trained and the model keeps updating weights from where is has left off.
    
    with open('sgd_partial_fit.pickle','wb') as file:
        pickle.dump(sgd_mini_batch_model_loaded,file)
```

```python
plt.plot(range(1,21),improvements_1)
plt.xticks(range(1,21))
plt.grid()
plt.ylabel('Test Accuracy')
plt.xlabel('Epochs')
```
