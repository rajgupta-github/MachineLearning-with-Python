```python
tips = sns.load_dataset("tips")
```

```python
tips.groupby('day')['total_bill'].mean().plot(kind='bar')
```

```python
sns.barplot(x="day", y="total_bill", estimator=sum, data=tips)
```

```python
sns.set_style("whitegrid")
sns.set_style("ticks")
sns.boxplot(x = "day", y = "total_bill", data = tips)
sns.despine()
```

```python
import numpy as np
```

```python
# This function shaLL help us plot some offset sine waves
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)

with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()

plt.subplot(212)
sinplot(-1)

```
